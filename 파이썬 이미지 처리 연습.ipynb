```python
from tkinter import *
from tkinter import messagebox
import math
import os.path
def rotateImage_45():
    global window, canvas, paper, height, width, filename
    canvas.destroy()
    rotated_image = [[0 for _ in range(width)] for _ in range(height)]
    center_x = width / 2
    center_y = height / 2
    theta = math.radians(45)  # 각도를 라디안으로 변환

    for i in range(height):
        for k in range(width):
            x = k - center_x
            y = i - center_y
            new_x = int(x * math.cos(theta) - y * math.sin(theta) + center_x)
            new_y = int(x * math.sin(theta) + y * math.cos(theta) + center_y)
            
            if 0 <= new_x < width and 0 <= new_y < height:
                rotated_image[i][k] = image[new_y][new_x]

    print(len(rotated_image))
    canvas = Canvas(window, height=256, width= 256)
    paper = PhotoImage(height = 256, width = 256)
    canvas.create_image((256/2, 256/2), image = paper, state='normal')
    for i in range(height):
        for k in range(width):
            r = g = b = rotated_image[i][k] # 3원색이 같으면 그레이 스케일
            paper.put('#%02x%02x%02x' % (r, g, b), (i, k))# rgb를 16진수로 찍겠다.
    canvas.pack()



# 함수
def displayImage():
    global window, canvas, paper, height, width, filename
    if canvas!=None:
        canvas.destroy()
    canvas = Canvas(window, height=256, width= 256)
    paper = PhotoImage(height = 256, width = 256)
    canvas.create_image((256/2, 256/2), image = paper, state='normal')
    for i in range(height):
        for k in range(width):
            r = g = b = image[i][k] # 3원색이 같으면 그레이 스케일
            paper.put('#%02x%02x%02x' % (r, g, b), (i, k))# rgb를 16진수로 찍겠다.
    canvas.pack()
def reverseImage():
    for i in range(height) :
        for k in range(width) :
            image[i][k] = 255 - image[i][k]
    displayImage()
def avgColor():
    s=0
    for i in range(height):
        s+=sum(image[i])
    avg = s/256**2
    
    for i in range(height):
        for k in range(width):
            if image[i][k]>=avg:
                image[i][k]=255
            else:
                image[i][k]=0
    displayImage()
def size2():#두배로~
    global window, canvas, paper, height, width, filename
    canvas.destroy()
    canvas = Canvas(window, height = 512, width=512)
    paper = PhotoImage(height=512, width=512)
    canvas.create_image((512/2, 512/2), image= paper, state='normal')
    for i in range(512):
        for k in range(512):
            r = g = b = image2[i][k] # 3원색이 같으면 그레이 스케일
            paper.put('#%02x%02x%02x' % (r, g, b), (i, k))# rgb를 16진수로 찍겠다.

    canvas.pack()

def sizeDown2():
    global window, canvas, paper, height, width, filename
    canvas.destroy()
    canvas = Canvas(window, height = 128, width=128)
    paper = PhotoImage(height=128, width=128)
    canvas.create_image((128/2, 128/2), image= paper, state='normal')
    print(width)
    image3 =[[] for _ in range(width//2)]
    for i in range(0,height,2):
        for j in range(0,width,2):
            s=0
            s+=image[i][j]
            s+=image[i][j+1]
            s+=image[i+1][j]
            s+=image[i+1][j+1]
            image3[i//2].append(s//4)
    print(len(image3), len(image3[0]))
    for i in range(128):
        for k in range(128):
            r = g = b = image3[i][k] # 3원색이 같으면 그레이 스케일
            paper.put('#%02x%02x%02x' % (r, g, b), (i, k))
    canvas.pack()

# 변수
window, canvas, paper = None, None, None
filename = ""
height, width = 0, 0
image = []


# 메인
filename = 'new.RAW'
# 파일 크기 알아내기
fSize = os.path.getsize(filename) # Byte 단위
height = width = int(math.sqrt(fSize))

image = [ [0 for _ in range(width)] for _ in range(height)]
rfp = open(filename, 'rb')
image2 = [[] for _ in range(height*2)]

for i in range(height-1,-1,-1) :
    for k in range(width-1,-1,-1) :
        image[i][k] = ord(rfp.read(1))
rfp.close()

rfp2 = open(filename, 'rb')
for i in range(0,height*2,2):
    for j in range(width):
        u= ord(rfp2.read(1))
        image2[i].append(u)
        image2[i].append(u)
        image2[i+1].append(u)
        image2[i+1].append(u)
rfp2.close()



# 절취선
window = Tk()
window.geometry('900x900')
window.title('영상처리 Alpha')


btnReverse = Button(window, text='반전', command = reverseImage)
btnGray = Button(window, text='흑백', command = avgColor)
btnSize2 = Button(window, text="두 배로 확대", command= size2)
btnSizeDown2 = Button(window, text="두 배로 축소", command = sizeDown2)
btn45r = Button(window, text = "45도 회전", command=rotateImage_45)

btn45r.pack()
btnReverse.pack()
btnGray.pack()
btnSize2.pack()
btnSizeDown2.pack()
displayImage()
window.mainloop()
```
