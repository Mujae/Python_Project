# 여기서부터 동영상 분석

```python
# import the opencv library
import cv2
import torch # 토치 추가
import numpy as np
import time

# define a video capture object

model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best_last.pt')  # local model
filepath = '/content/apsd.mp4'
video = cv2.VideoCapture(filepath) #'' 사이에 사용할 비디오 파일의 경로 및 이름을 넣어주도록 함

if not video.isOpened():
    print("Could not Open :", filepath)
    exit(0)
    
#불러온 비디오 파일의 정보 출력
length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

fourcc=cv2.VideoWriter_fourcc('m', 'p', '4', 'v') #코덱 정보
writer=cv2.VideoWriter('/content/go2/output_last.mp4', fourcc, fps, (width, height)) #동영상 저장을 위한 Writer

blue_color=(255,0,0)
other_color=(0, 0, 255)
best_t=10000 # 반응속도 제일 빨랐던 시간
cc_on = 0 #cc기 켜진건지
stime = 0 #시작 시간
t_time=100
cleanse=0
while(video.isOpened()): # length가 비디오 전체 길이였으니 frame_index를 하나씩 올리다가 length가 되면 탈출
  ret, frame = video.read()
  if not ret:#ret에 True가 들어오면 제대로 값이 들어온 것이고 false면 로딩실패기 때문에 탈출
    break
  results=model(frame) # 만든 모델로 frame 분석!
  results_rf = results.pandas().xyxy[0].values
  for a in results_rf:#한 프레임에 여러개 있을 수도 있어서 for문으로
    if (cc_on==0 and a[4]>0.4) and (int(a[6])>2 and int(a[6])<14): #현재 프레임에서 cc기가
      fr = cv2.rectangle(frame, (int(a[0]), int(a[1])), (int(a[2]), int(a[3])), blue_color, 3)
      cv2.putText(frame, "CC기"+str(a[4]), (int(a[0]), int(a[1])), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (36,255,12), 2)
      cc_on=1
      stime = time.time()
      print(str(a[4]), a[5], 1)
      continue
    if (int(a[6])==1 or int(a[6])==2) and (cleanse==0 and a[4]>0.4) and cc_on==1:
      fr = cv2.rectangle(frame, (int(a[0]), int(a[1])), (int(a[2]), int(a[3])), other_color, 3)
      cv2.putText(frame, "cleanse"+str(a[4]), (int(a[0]), int(a[1])), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (36,255,12), 2)
      af=time.time()
      t_time=min(af-stime,t_time)
      cleanse=1
      print(str(a[4]), a[5], 2)

  writer.write(frame)
    

# After the loop release the cap object
video.release()
writer.release()

print("length :", length)
print("width :", width)
print("height :", height)
print("fps :", fps)
if cleanse==0:
  print("결과: 실패")
else:
  print(t_time, end=" ")

  if t_time<=0.15:
    print("0.15초 이하.. 당신은 프로게이머?")
  elif t_time<=0.5:
    print("0.5초 이하.. 준수한 반응속도군요.")
  elif t_time<=1.0:
    print("1초 이하.. 정상인 범주입니다.")
  elif t_time<=2.0:
    print("당신은 벌레입니다.")
  else:
    print("수은이랑 정화 왜 씀?ㅋㅋ")



# Destroy all the windows
cv2.destroyAllWindows()
```
