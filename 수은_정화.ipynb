```python
# import the opencv library
import cv2
import torch # í† ì¹˜ ì¶”ê°€
import numpy as np
import time

# define a video capture object

model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best_last.pt')  # local model
filepath = '/content/apsd.mp4'
video = cv2.VideoCapture(filepath) #'' ì‚¬ì´ì— ì‚¬ìš©í•  ë¹„ë””ì˜¤ íŒŒì¼ì˜ ê²½ë¡œ ë° ì´ë¦„ì„ ë„£ì–´ì£¼ë„ë¡ í•¨

if not video.isOpened():
    print("Could not Open :", filepath)
    exit(0)
    
#ë¶ˆëŸ¬ì˜¨ ë¹„ë””ì˜¤ íŒŒì¼ì˜ ì •ë³´ ì¶œë ¥
length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

fourcc=cv2.VideoWriter_fourcc('m', 'p', '4', 'v') #ì½”ë± ì •ë³´
writer=cv2.VideoWriter('/content/go2/output_last.mp4', fourcc, fps, (width, height)) #ë™ì˜ìƒ ì €ìž¥ì„ ìœ„í•œ Writer

blue_color=(255,0,0)
other_color=(0, 0, 255)
best_t=10000 # ë°˜ì‘ì†ë„ ì œì¼ ë¹¨ëžë˜ ì‹œê°„
cc_on = 0 #ccê¸° ì¼œì§„ê±´ì§€
stime = 0 #ì‹œìž‘ ì‹œê°„
t_time=100
cleanse=0
while(video.isOpened()): # lengthê°€ ë¹„ë””ì˜¤ ì „ì²´ ê¸¸ì´ì˜€ìœ¼ë‹ˆ frame_indexë¥¼ í•˜ë‚˜ì”© ì˜¬ë¦¬ë‹¤ê°€ lengthê°€ ë˜ë©´ íƒˆì¶œ
  ret, frame = video.read()
  if not ret:#retì— Trueê°€ ë“¤ì–´ì˜¤ë©´ ì œëŒ€ë¡œ ê°’ì´ ë“¤ì–´ì˜¨ ê²ƒì´ê³  falseë©´ ë¡œë”©ì‹¤íŒ¨ê¸° ë•Œë¬¸ì— íƒˆì¶œ
    break
  results=model(frame) # ë§Œë“  ëª¨ë¸ë¡œ frame ë¶„ì„!
  results_rf = results.pandas().xyxy[0].values
  for a in results_rf:#í•œ í”„ë ˆìž„ì— ì—¬ëŸ¬ê°œ ìžˆì„ ìˆ˜ë„ ìžˆì–´ì„œ forë¬¸ìœ¼ë¡œ
    if (cc_on==0 and a[4]>0.4) and (int(a[6])>2 and int(a[6])<14): #í˜„ìž¬ í”„ë ˆìž„ì—ì„œ ccê¸°ê°€
      fr = cv2.rectangle(frame, (int(a[0]), int(a[1])), (int(a[2]), int(a[3])), blue_color, 3)
      cv2.putText(frame, "CCê¸°"+str(a[4]), (int(a[0]), int(a[1])), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (36,255,12), 2)
      cc_on=1
      stime = time.time()
      print(str(a[4]), a[5], 1)
      continue
    if (int(a[6])==1 or int(a[6])==2) and (cleanse==0 and a[4]>0.4) and cc_on==1:
      fr = cv2.rectangle(frame, (int(a[0]), int(a[1])), (int(a[2]), int(a[3])), other_color, 3)
      cv2.putText(frame, "cleanse"+str(a[4]), (int(a[0]), int(a[1])), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (36,255,12), 2)
      af=time.time()
      t_time=min(af-stime,t_time)
      cleanse=1
      print(str(a[4]), a[5], 2)

  writer.write(frame)
    

# After the loop release the cap object
video.release()
writer.release()

print("length :", length)
print("width :", width)
print("height :", height)
print("fps :", fps)
if cleanse==0:
  print("ê²°ê³¼: ì‹¤íŒ¨")
else:
  print(t_time, end=" ")

  if t_time<=0.15:
    print("0.15ì´ˆ ì´í•˜.. ë‹¹ì‹ ì€ í”„ë¡œê²Œì´ë¨¸?")
  elif t_time<=0.5:
    print("0.5ì´ˆ ì´í•˜.. ì¤€ìˆ˜í•œ ë°˜ì‘ì†ë„êµ°ìš”.")
  elif t_time<=1.0:
    print("1ì´ˆ ì´í•˜.. ì •ìƒì¸ ë²”ì£¼ìž…ë‹ˆë‹¤.")
  elif t_time<=2.0:
    print("ë‹¹ì‹ ì€ ë²Œë ˆìž…ë‹ˆë‹¤.")
  else:
    print("ìˆ˜ì€ì´ëž‘ ì •í™” ì™œ ì”€?ã…‹ã…‹")



# Destroy all the windows
cv2.destroyAllWindows()
```
